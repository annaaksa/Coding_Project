```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import pandas as pd

# Read data from the file
data = pd.read_csv('data5-1.csv')

# Create histogram
plt.hist(data, bins=30, density=True, alpha=0.6, color='g')

# Fit a normal distribution to the data
mu, std = norm.fit(data)

# Plot the PDF
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2)

# Calculate mean annual salary (W~)
mean_salary = np.mean(data)

# Calculate X such that 10% of people have a salary below X
X = np.percentile(data, 10)

# Plot mean and X on the graph
plt.axvline(mean_salary, color='r', linestyle='dashed', linewidth=2, label=f'Mean Salary: {mean_salary:.2f}')
plt.axvline(X, color='b', linestyle='dashed', linewidth=2, label=f'X (10% below): {X:.2f}')

# Labeling and title
plt.xlabel('Annual Salary')
plt.ylabel('Probability Density')
plt.title('Probability Density Function of Annual Salaries')
plt.legend()

# Show the plot  
plt.show()

# Print calculated values
print(f"Mean Annual Salary (W~): {mean_salary:.2f}")
print(f"Value X (10% below): {X:.2f}")
```
